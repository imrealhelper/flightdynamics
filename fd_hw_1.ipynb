```python
# Split the dataframe into 7 smaller dataframes based on the AOA values -9, -6, -3, 0, 3, 6, and 9
columns_per_section = 5  # Each section has 5 columns (AOA, CL, CD, Cm, and NaN)
sections = [-9, -6, -3, 0, 3, 6, 9]

dfs = {}
for i, aoa in enumerate(sections):
    start_col = i * columns_per_section
    end_col = start_col + columns_per_section
    dfs[aoa] = df.iloc[:, start_col:end_col]

for i in dfs:
  display(dfs[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Assuming c_gs and dfs are defined
C_l_d_x = {}
C_l_d_y = {}

for k in range(len(c_gs)):
    C_l_d_y[int(100 * k)] = []
    C_l_d_x[int(100 * k)] = []
    for i in dfs:
        C_M_array = np.array(list((dfs[i]["Cm"] + dfs[i]["CL"] * (c_gs[k] - 0.25))))
        C_L_array = list(dfs[i]["CL"])
        C_D_array = list(dfs[i]["CD"])
        d = i  # `d` is now the key `i` in `dfs`

        positive_index = (np.where(C_M_array > 0, C_M_array, np.inf)).argmin()  # Closest positive
        negative_index = (np.where(C_M_array < 0, C_M_array, -np.inf)).argmax()  # Closest negative

        positive_distance = C_M_array[positive_index]
        negative_distance = -C_M_array[negative_index]
        distance = positive_distance + negative_distance

        # Interpolated lift coefficient (CL) at deflection angle (d)
        interpolated_CL = (C_L_array[positive_index] - C_L_array[negative_index]) * (negative_distance / distance) + C_L_array[negative_index]

        C_l_d_x[int(100 * k)].append(d)  # Store `d` values (the keys of dfs)
        C_l_d_y[int(100 * k)].append(interpolated_CL)  # Corresponding interpolated CL values

# Plot with interpolation for a smooth curve
for k in range(len(c_gs)):
    x_vals = C_l_d_x[int(100 * k)]
    y_vals = C_l_d_y[int(100 * k)]

    # Trim the last point if needed
    x_vals = x_vals[:len(x_vals)-1]
    y_vals = y_vals[:len(y_vals)-1]

    # Interpolation for smoothness
    if len(x_vals) > 1:  # Ensure there are enough points for interpolation
        x_new = np.linspace(min(x_vals), max(x_vals), 10000)  # 500 points for a smooth curve
        f_interpolated = interp1d(x_vals, y_vals, kind='cubic')  # Cubic spline interpolation
        y_smooth = f_interpolated(x_new)
        x_new = x_new[y_smooth<1.2]
        y_smooth = y_smooth[y_smooth<1.2]
        plt.plot(x_new, y_smooth, label=f"MRC: {int(100*c_gs[k])}%")

plt.xlabel("($d$)")
plt.ylabel("($C_L$)")
plt.legend()
plt.grid()
plt.show()

```
